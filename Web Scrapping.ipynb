Question no 9

```python
rest = requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
rest
```

```python
soup2 = BeautifulSoup(rest.content)
```

```python
restname = []
for i in soup2.find_all('a', class_="restnt-name ellipsis"):
    restname.append(i.text)
```

```python
restname
```

```python
cuisine = []
for i in soup2.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
    
```

```python
cuisine
```

```python
location = []
for i in soup2.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)
```

```python
location
```

```python
ratings = []
for i in soup2.find_all('div', class_="restnt-rating rating-4"):
    ratings.append(i.text)
```

```python
ratings
```

```python
images = []
for i in soup2.find_all('img', class_="no-img"):
    images.append(i.get('data-src'))
```

```python
images
```

```python
import pandas as pd
```

```python
print(len(restname),len(cuisine),len(location),len(ratings),len(images))
```

```python
DF = pd.DataFrame({'Restaurant Name': restname,'Cuisine': cuisine,'Location':location,'Ratings':ratings,'Images URL':images})
DF
```
